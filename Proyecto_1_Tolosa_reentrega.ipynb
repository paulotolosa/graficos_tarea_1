```python
import pandas as pd

# URL directa al archivo CSV en GitHub
url = 'https://raw.githubusercontent.com/paulotolosa/graficos_tarea_1/main/diabetes.csv'

# Importar el archivo CSV en un DataFrame de pandas
df = pd.read_csv(url)

# Ver las primeras filas del DataFrame para asegurarse de que se importó correctamente
print(df.head())
```

Histograma de Glucosa:
Puedes trazar un histograma para ver la distribución de los niveles de glucosa en la población del estudio.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Suponiendo que 'df' es tu DataFrame con los datos de diabetes

# 1. Histograma de Glucosa
plt.figure(figsize=(8, 6))
plt.hist(df['Glucose'], bins=20, color='skyblue')
plt.xlabel('Niveles de Glucosa')
plt.ylabel('Frecuencia')
plt.title('Histograma de Niveles de Glucosa')
plt.grid(True)
plt.show()
```

Diagrama de dispersión de BMI vs. Edad:
Un diagrama de dispersión puede mostrarte si hay alguna relación entre el índice de masa corporal (BMI) y la edad de los pacientes.

```python
# 2. Diagrama de dispersión de BMI vs. Edad
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['BMI'], color='orange', alpha=0.5)
plt.xlabel('Edad')
plt.ylabel('BMI')
plt.title('BMI vs. Edad')
plt.grid(True)
plt.show()
```

Gráfico de cajas de Presión Sanguínea (BloodPressure):
Un gráfico de cajas puede mostrar la distribución de la presión sanguínea y los valores atípicos.

```python
import seaborn as sns

# 3. Gráfico de cajas de Presión Sanguínea (BloodPressure)
plt.figure(figsize=(8, 6))
plt.boxplot(df['BloodPressure'])
plt.xlabel('Presión Sanguínea')
plt.title('Distribución de Presión Sanguínea')
plt.grid(True)
plt.show()
```

Diagrama de dispersión con 'Outcome' como hue:
Este gráfico muestra un diagrama de dispersión con diferentes colores para cada valor de 'Outcome', lo que permite visualizar la relación entre dos variables con respecto al resultado.

```python
import seaborn as sns

# 1. Diagrama de dispersión con 'Outcome' como hue
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=df)
plt.title('Diagrama de Dispersión con Outcome como hue')
plt.xlabel('Glucosa')
plt.ylabel('BMI')
plt.grid(True)
plt.show()
```

Comparación la distribución de la glucosa, la presión sanguínea y el índice de masa corporal (BMI) entre las personas con y sin diabetes ('Outcome'). Podemos crear un grid de 1 fila y 3 columnas para colocar cada gráfico en una columna diferente.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Suponiendo que 'df' es tu DataFrame con los datos

# Crear un grid de 1 fila y 3 columnas
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Gráfico 1: Distribución de Glucosa
sns.histplot(df, x='Glucose', hue='Outcome', kde=True, ax=axes[0])
axes[0].set_title('Distribución de Glucosa')
axes[0].set_xlabel('Glucosa')

# Gráfico 2: Distribución de Presión Sanguínea
sns.histplot(df, x='BloodPressure', hue='Outcome', kde=True, ax=axes[1])
axes[1].set_title('Distribución de Presión Sanguínea')
axes[1].set_xlabel('Presión Sanguínea')

# Gráfico 3: Distribución de BMI
sns.histplot(df, x='BMI', hue='Outcome', kde=True, ax=axes[2])
axes[2].set_title('Distribución de BMI')
axes[2].set_xlabel('BMI')

# Ajustar el espacio entre los subgráficos
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```
