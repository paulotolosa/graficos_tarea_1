```python
import pandas as pd

# URL directa al archivo CSV en GitHub
url = 'https://raw.githubusercontent.com/paulotolosa/graficos_tarea_1/main/diabetes.csv'

# Importar el archivo CSV en un DataFrame de pandas
df = pd.read_csv(url)

# Ver las primeras filas del DataFrame para asegurarse de que se importó correctamente
print(df.head())
```

Histograma de Glucosa:
Puedes trazar un histograma para ver la distribución de los niveles de glucosa en la población del estudio.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Suponiendo que 'df' es tu DataFrame con los datos de diabetes

# 1. Histograma de Glucosa
plt.figure(figsize=(8, 6))
plt.hist(df['Glucose'], bins=20, color='skyblue')
plt.xlabel('Niveles de Glucosa')
plt.ylabel('Frecuencia')
plt.title('Histograma de Niveles de Glucosa')
plt.grid(True)
plt.show()
```

Diagrama de dispersión de BMI vs. Edad:
Un diagrama de dispersión puede mostrarte si hay alguna relación entre el índice de masa corporal (BMI) y la edad de los pacientes.

```python
# 2. Diagrama de dispersión de BMI vs. Edad
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['BMI'], color='orange', alpha=0.5)
plt.xlabel('Edad')
plt.ylabel('BMI')
plt.title('BMI vs. Edad')
plt.grid(True)
plt.show()
```

Gráfico de cajas de Presión Sanguínea (BloodPressure):
Un gráfico de cajas puede mostrar la distribución de la presión sanguínea y los valores atípicos.

```python
# 3. Gráfico de cajas de Presión Sanguínea (BloodPressure)
plt.figure(figsize=(8, 6))
plt.boxplot(df['BloodPressure'])
plt.xlabel('Presión Sanguínea')
plt.title('Distribución de Presión Sanguínea')
plt.grid(True)
plt.show()
```

Diagrama de dispersión con regresión lineal y 'Outcome' como hue:
Este gráfico mostrará un diagrama de dispersión con una regresión lineal para cada valor de 'Outcome', diferenciado por color.

```python
import seaborn as sns

# 1. Diagrama de dispersión con regresión lineal y 'Outcome' como hue
plt.figure(figsize=(8, 6))
sns.lmplot(x='Glucose', y='BMI', hue='Outcome', data=df, fit_reg=True)
plt.title('Diagrama de Dispersión con Regresión Lineal (hue=Outcome)')
plt.xlabel('Glucosa')
plt.ylabel('BMI')
plt.grid(True)
plt.show()
```

Gráfico de barras con 'Outcome' como hue y grid:
Un gráfico de barras que muestra la distribución de una variable categórica en función de otra variable categórica ('Outcome'), con subdivisiones en el grid.

```python
# 2. Gráfico de barras con 'Outcome' como hue y grid
plt.figure(figsize=(10, 8))
sns.countplot(x='Pregnancies', hue='Outcome', data=df)
plt.title('Gráfico de Barras de Embarazos con Grid (hue=Outcome)')
plt.xlabel('Número de Embarazos')
plt.ylabel('Recuento')
plt.grid(True)
plt.show()
```

Diagrama de cajas con 'Outcome' como hue y grid:
Este gráfico mostrará la distribución de una variable numérica en diferentes valores de 'Outcome', separados por columnas en el grid.

```python
# 3. Diagrama de cajas con 'Outcome' como hue y grid
plt.figure(figsize=(10, 8))
sns.boxplot(x='Outcome', y='Age', hue='Outcome', data=df)
plt.title('Diagrama de Cajas de Edad con Grid (hue=Outcome)')
plt.xlabel('Resultado')
plt.ylabel('Edad')
plt.grid(True)
plt.show()
```
